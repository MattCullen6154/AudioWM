## Robust Watermarking for Telephone Audio Authentication

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import scipy.fft as fft
import soundfile as sf
from IPython.display import Audio, display
from pathlib import Path
import librosa
import librosa.display

plt.rcParams["figure.figsize"] = (10, 4)
plt.rcParams["axes.grid"] = True

RNG = np.random.default_rng(12345)
```

```python

AUDIO_PATH = "data/Cortes2.wav"

# Framing
FRAME_MS = 20          # frame length in ms
HOP_MS = 10            # hop length in ms
NFFT = 1024

# Watermarking
EMBED_STRENGTH = 0.05
BIT_REPEAT = 4
WATERMARK_BITS = np.array([1, 0, 1, 1, 0, 0, 1, 0], dtype=int)

# Frequency bin selection
FREQ_LOW = 600         # Hz
FREQ_HIGH = 2500       # Hz

# Distortion toggles
APPLY_BANDLIMIT = True
APPLY_AWGN = True
APPLY_RESAMPLE = False
APPLY_QUANTIZATION = False

# Noise level
NOISE_STD = 0.003
```

```python
def load_audio(path):
    x, fs = sf.read(path)
    if x.ndim > 1:
        x = np.mean(x, axis=1)  # convert stereo to mono
    return x.astype(np.float64), fs

def normalize_audio(x, peak=0.99):
    m = np.max(np.abs(x))
    if m < 1e-12:
        return x.copy()
    return peak * x / m

def to_float64(x):
    return np.asarray(x, dtype=np.float64)

    
```

```python
def ms_to_samples(ms, fs):
    return int(round(ms * fs / 1000))

def frame_signal(x, frame_length, hop_length):
    num_frames = 1 + max(0, (len(x) - frame_length) // hop_length)
    frames = np.stack([
        x[i * hop_length : i * hop_length + frame_length]
        for i in range(num_frames)
    ], axis=0)
    return frames

def overlap_add(frames, hop_length, window=None):
    num_frames, frame_length = frames.shape
    out_len = (num_frames - 1) * hop_length + frame_length
    y = np.zeros(out_len)
    wsum = np.zeros(out_len)

    if window is None:
        window = np.ones(frame_length)

    for i, frame in enumerate(frames):
        start = i * hop_length
        y[start:start + frame_length] += frame * window
        wsum[start:start + frame_length] += window**2

    nz = wsum > 1e-10
    y[nz] /= wsum[nz]
    return y
```

```python
def get_window(frame_length):
    return signal.windows.hann(frame_length, sym=False)

def compute_fft_bins(fs, nfft):
    return np.fft.rfftfreq(nfft, d=1/fs)

def get_bin_mask(freqs, f_low, f_high):
    return (freqs >= f_low) & (freqs <= f_high)

def stft_frames(x, fs, frame_ms=20, hop_ms=10, nfft=512):
    frame_len = ms_to_samples(frame_ms, fs)
    hop_len = ms_to_samples(hop_ms, fs)
    w = get_window(frame_len)

    frames = frame_signal(x, frame_len, hop_len)
    windowed = frames * w[None, :]
    X = np.fft.rfft(windowed, n=nfft, axis=1)
    freqs = compute_fft_bins(fs, nfft)

    return X, freqs, frame_len, hop_len, w
```

```python
def plot_waveform(x, fs, title="Waveform"):
    t = np.arange(len(x)) / fs
    plt.figure()
    plt.plot(t, x)
    plt.title(title)
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()

def plot_full_spectrum(x, fs=1.0, title="Spectrum of Full Signal"):
    N = len(x)
    X = np.fft.fft(x)

    X_mag = np.abs(X)[:N//2]          # magnitude
    freqs = np.fft.fftfreq(N, d=1/fs)[:N//2]   # real frequency axis
    
    plt.figure(figsize=(10, 4))
    plt.plot(freqs, X_mag)
    plt.title(title)
    plt.xlabel("Frequency [Hz]")
    plt.ylabel("Magnitude")
    plt.xlim(0, 4000)  
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_spectrum_of_frame(X_frame, freqs, title="Frame Spectrum"):
    plt.figure()
    plt.plot(freqs, np.abs(X_frame))
    plt.title(title)
    plt.xlabel("Frequency [Hz]")
    plt.ylabel("Magnitude")
    plt.show()

def plot_spectrogram(x, fs, title="Spectrogram"):
    f, t, Sxx = signal.spectrogram(x, fs=fs, nperseg=256, noverlap=128)
    plt.figure()
    plt.pcolormesh(t, f, 10*np.log10(Sxx + 1e-12), shading="gouraud")
    plt.title(title)
    plt.ylabel("Frequency [Hz]")
    plt.xlabel("Time [s]")
    plt.colorbar(label="Power [dB]")
    plt.ylim(0, 4000)
    plt.show()
```

```python
# Load Speech Audio
input_audio, fs = load_audio(AUDIO_PATH)
input_normalized = normalize_audio(input_audio)
print(f"Sample rate: {fs} Hz")
print(f"Duration: {len(input_normalized)/fs:.2f} s")

display(Audio(input_normalized, rate=fs))
plot_waveform(input_normalized, fs, "Original Speech")
plot_full_spectrum(input_normalized, fs, "Spectrum of Full Signal")
plot_spectrogram(input_normalized, fs, "Original Speech Spectrogram")
```
